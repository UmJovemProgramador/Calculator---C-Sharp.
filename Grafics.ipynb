```python
import matplotlib.pyplot as plt
x = ["Camisa", "Short", "Tênis", "Sandália"]
y = [8, 3, 5, 12]
fig, grafico = plt.subplots()
grafico.plot(x, y, color ='red',linestyle='--',marker='*')
grafico.set_title('Vendas')
grafico.set_xlabel('Produtos')
grafico.set_ylabel('Quantidade')
```

```python
import matplotlib.pyplot as plt
x = ["Camisa", "Short", "Tênis", "Sandália"]
y = [8, 3, 5, 12]
y2 = [60.50, 25.37, 40, 120.50]
fig, grafico = plt.subplots()
grafico.plot(x,y,color='green', 
             linestyle='--',marker='o', 
             label='Quantidade')
grafico.plot(x,y2,color='blue',marker='*',
             label='Faturamento')
grafico.set_title('Vendas')
grafico.set_xlabel('Produtos')
grafico.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
times = ['Vontade de comer miojo', 'Botafogo', 'Paysandu', 'Palmeiras']
torcedores = [,12,6,4]
cores = ['yellow', 'green', 'blue', 'orange']
fig, grafico = plt.subplots()
grafico.pie(torcedores, labels = times,
            colors=cores, autopct='%1.1f%%')
grafico.set_title("% de torcedores")
plt.show()
```
